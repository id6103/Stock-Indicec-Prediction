```python
import pandas as pd
import numpy as np
    
from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
sid = SentimentIntensityAnalyzer()



for i in range(0,len(news_final)):
    for j in range(0,10):
    
      daily_news = news_final['submissions'][i][j]['title']
      ss = sid.polarity_scores(daily_news)
      date = news_final['startAt'][i]
      score = news_final['submissions'][i][j]['score']
      num_comments = news_final['submissions'][i][j]['num_comments']
      neg = float(ss['neg'])
      neu = ss['neu']
      pos = ss['pos']
      compound = ss['compound']
      sentiment_data14.loc[len(sentiment_data14)]=[str(date), score, num_comments,neg,neu,pos,compound]
      

# sentiment_data13




```

```python
sentiment_data14.to_csv("sentiment_data14.csv", index = False)
```

```python
import pandas as pd

news_final = pd.read_json("newsFinal14.json")
```

```python
# TF - IDF
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

content = []

# for i in range(0,l):
for i in range(0,len(news_final)):
    for j in range(0,10):
      content.append(news_final['submissions'][i][j]['title'])
      
# print (content)
tvz = TfidfVectorizer(max_features = 200, max_df = 0.8,sublinear_tf=True, use_idf =True, stop_words = 'english')
X = tvz.fit_transform(content)
feature_names = tvz.get_feature_names()
dense = X.todense()
denselist = dense.tolist()
tfidf14 = pd.DataFrame(denselist, columns=feature_names)
```

```python
tfidf14.to_csv("tfidf14.csv", index = False)
```

```python
import pandas as pd

sentiment_data14 = pd.read_csv("sentiment_data14.csv")
tfidf14 = pd.read_csv("tfidf14.csv")
sentiment_tfidf14 = pd.concat([sentiment_data14, tfidf14], axis=1)
sentiment_tfidf14.to_csv("sentiment_tfidf14.csv", index = False) 
```

```python
import time

mean_sentiment_tfidf14 = sentiment_tfidf14.groupby(['date'], as_index=False).mean()
mean_sentiment_tfidf14 = mean_sentiment_tfidf14.assign()
mean_sentiment_tfidf14.to_csv("mean_sentimentAidf14.csv", index = False) 
```

```python
#convert epoch to date yyyy-mm-dd

for k in range(0, len(mean_sentiment_tfidf14)):
    mean_sentiment_tfidf14['date'][k]= time.strftime('%Y-%m-%d', time.localtime(int(mean_sentiment_tfidf14['date'][k])))


mean_sentiment_tfidf14.to_csv("mean_sentiment_tfidf14.csv", index = False)
```
