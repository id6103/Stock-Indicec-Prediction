```python
#sort with time desc
news_final = news_final.iloc[::-1]
news_final.index = range(len(news_final))
len(news_final)
```

```python
import nltk
nltk.download("all")
```

```python
# print (news_final['submissions'][0])
# print (news_final['submissions'][0][0])
# print (news_final['submissions'][0][0]['title'])

print (news_final['submissions'][0][1]['title'])
sentiment_data14 = pd.DataFrame(columns=(['date','score','num_comments','neg','neu','pos','compound']))
sentiment_data14
```

```python
import pandas as pd
import numpy as np
    
from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
sid = SentimentIntensityAnalyzer()



for i in range(0,len(news_final)):
    for j in range(0,10):
    
      daily_news = news_final['submissions'][i][j]['title']
      ss = sid.polarity_scores(daily_news)
      date = news_final['startAt'][i]
      score = news_final['submissions'][i][j]['score']
      num_comments = news_final['submissions'][i][j]['num_comments']
      neg = float(ss['neg'])
      neu = ss['neu']
      pos = ss['pos']
      compound = ss['compound']
      sentiment_data14.loc[len(sentiment_data14)]=[str(date), score, num_comments,neg,neu,pos,compound]
      

sentiment_data14
# for i in range(len(news)):
#     sentence = news['title'][i]
#     print(sentence)
#     ss = sid.polarity_scores(sentence)
# #         print (ss['neg'])
#     rating_data.iloc[i, 1] = float(ss['neg'])
# #         print (rating_data.iloc[i, 1])
#     rating_data.iloc[i, 2] = ss['neu']
#     rating_data.iloc[i, 3] = ss['pos']
#     rating_data.iloc[i, 4] = ss['compound']



```

```python
sentiment_data14.to_excel("sentiment_data14.xlsx", index = False)
```

```python
import time

mean_sentiment = sentiment_data14.groupby(['date'], as_index=False).mean()
mean_sentiment = mean_sentiment.assign()

mean_sentiment14 = mean_sentiment.copy()

for k in range(0, len(mean_sentiment)):
    mean_sentiment14['date'][k]= time.strftime('%Y-%m-%d', time.localtime(int(mean_sentiment['date'][k])))

mean_sentiment14
```
